```
## use Pandas to read 圖片
import pandas as pd
import glob

df = pd.DataFrame(columns=["path", "target"])
for fn in glob.glob("./Tshirt_*.jpg"):
    s = pd.Series([fn, 0], index=["path", "target"])
    df = df.append(s, ignore_index=True)
for fn in glob.glob("./shirt_*.jpg"):
    s = pd.Series([fn, 1], index=["path", "target"])
    df = df.append(s, ignore_index=True)
df
```

```
import random
from keras.preprocessing.image import load_img
c = random.randint(0, 938)
print("Tshirt or shirt?", df.iloc[c]["target"])
img = load_img(df.iloc[c]["path"])
print(img.size)
img
```

```
img = load_img(df.iloc[c]["path"], target_size=(224, 224))
print(img.size)
img
```

```
from keras.applications.vgg16 import VGG16
vgg = VGG16(include_top=False, input_shape=(224, 224, 3))
vgg.summary()
```

```
vgg.layers
```

```
from keras.models import Model
from keras.layers import Dense, Flatten, BatchNormalization
for l in vgg.layers:
    l.trainable = False
# 重點中的重點中的重點
x = BatchNormalization()(vgg.output)
x = Flatten()(x)
x = Dense(1024, activation="relu")(x)
x = Dense(128, activation="relu")(x)
# 第一個選擇: 1 activation="sigmoid"
# 第二個選擇: 2 activation="softmax"
x = Dense(2, activation="softmax")(x)
cnn = Model(inputs=vgg.input, outputs=x)
cnn.summary()
```

```
cnn.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```
# 在你開始訓練前, 所有東西都先轉np array
from sklearn.model_selection import train_test_split
import numpy as np
x = np.array(df["path"])
y = np.array(df["target"])
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```
from keras.utils.np_utils import to_categorical
y_train_Tshirt = to_categorical(y_train)
y_test_Tshirt = to_categorical(y_test)
y_test_Tshirt[0]
```

```
from keras.applications.vgg16 import preprocess_input
batch = 10
train = 10
for i in range(train):
    # 如果是NUMPY的randint, 第二個數是不包刮
    idx = np.random.randint(0, x_train.shape[0], size=batch)
    imglist = []
    for fn in x_train[idx]:
        img = load_img(fn, target_size=(224, 224))
        imglist.append(preprocess_input(np.array(img)))
    xs = np.array(imglist)
    ys = y_train_Tshirt[idx]
    result = cnn.train_on_batch(xs, ys)
    print("-" * 15, i, "-" * 15)
    print(result)
```

```
# 如果是NUMPY的randint, 第二個數是不包刮
idx = np.random.randint(0, x_test.shape[0], size=100)
imglist = []
for fn in x_test[idx]:
    img = load_img(fn, target_size=(224, 224))
    imglist.append(preprocess_input(np.array(img)))
xs = np.array(imglist)
ys = y_test_Tshirt[idx]
cnn.evaluate(xs, ys)
```

```
cnn.predict(xs)
```

```
# 作圖(參考)
pre = cnn.predict(xs).argmax(axis=-1)
# 拿出預測標籤最大值的索引
pre
```

```
# 拿出正確答案最大值的索引
ys.argmax(axis=-1)
```

```
import matplotlib.pyplot as plt
%matplotlib inline
idx = np.nonzero(pre != ys.argmax(axis=-1))[0]
width = 5
height = int(len(idx) / 5) + 1
plt.figure(figsize=(14, 6))
for (i, img) in enumerate(np.array(imglist)[idx]):
    plt.subplot(height, width, i + 1)
    t = "[P]:" + str(pre[idx][i]) + "[O]:" + str(ys.argmax(axis=-1)[idx][i])
    plt.title(t)
    plt.axis("off")
    plt.imshow(img)
```
